```python
SHOPS = [
    "https://kith.com",
    "https://undefeated.com",
    "https://en.afew-store.com",
    "https://sneakerpolitics.com",
    "https://www.footkorner.com",
]
```

```python
OUT_DIR = "data/raw"
os.makedirs(OUT_DIR, exist_ok=True)
```

```python
def clean_shop_name(url: str) -> str:
    return re.sub(r"^https?://(www\.)?", "", url).split("/")[0]

def run_all():
    for base_url in SHOPS:
        shop = clean_shop_name(base_url)
        print(f"\n=== 🛒 {shop} ===")
        try:
            parser = ShopifyParser(base_url, rps=1.0, burst=3)
            products = parser.run(limit=1)
            out_path = os.path.join(OUT_DIR, f"{shop}.jsonl")
            with open(out_path, "w", encoding="utf-8") as f:
                for p in products:
                    f.write(json.dumps(p, ensure_ascii=False) + "\n")
            print(f"✅ {shop}: сохранено {len(products)} товаров в {out_path}")
        except Exception as e:
            print(f"❌ {shop}: ошибка — {e}")
```

```python
run_all()
```
