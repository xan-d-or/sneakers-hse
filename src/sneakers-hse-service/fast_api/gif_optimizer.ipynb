```python
from PIL import Image

im = Image.open('/Users/a.r.makarenko/Downloads/Запись-экрана-2026-01-11-в-18.03.44.gif')
print(im.is_animated)  # True for animated GIFs
print(im.n_frames)     # Total frames
```

```python
keep_frames = []  # List of Image objects
for i in range(im.n_frames):
    im.seek(i)
    if i % 10 == 0:  # Example: drop odd frames
        keep_frames.append(im.copy())
```

```python
im.info.get('duration', 100)
```

```python
keep_frames[0].save('output10.gif', save_all=True, append_images=keep_frames[1:], 
                    duration=im.info.get('duration', 100) * 10, loop=0, optimize=True)
```
